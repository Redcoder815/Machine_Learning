```python
import numpy as np
import matplotlib.pyplot as plt

class LinearRegression:
  def __init__(self,lr = 0.01, iteratons = 1000):
    self.lr = lr
    self.iterations = iteratons
    self.m = 0
    self.b = 0

  def fit(self, X, y):
    n = len(X)
    for _ in range(self.iterations):
      y_pred = self.m * X + self.b

      dm = (- 2 / n) * sum(X * (y - y_pred))
      db = (-2/n) * sum(y-y_pred)

      self.m = self.m - (self.lr * dm)
      self.b = self.b - (self.lr * db)

  def predictions(self, X):
    return self.m * X + self.b

# 1. Generate Sample Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# 2. Train Model
model = LinearRegression(lr=0.01, iteratons=1000)
model.fit(X, y)
predictions = model.predictions(X)

# 3. Visualization
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, predictions, color='red', label='Regression Line')
plt.title("Linear Regression from Scratch")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy as np

def gradient_descent(X, y, learning_rate=0.01, iterations=1000):
    m = 0.0  # Initialize slope
    b = 0.0  # Initialize intercept
    n = len(X)

    for i in range(iterations):
        # 1. Make predictions
        y_pred = m * X + b

        # 2. Compute gradients
        dm = (-2/n) * np.sum(X * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)

        # 3. Update parameters
        m = m - learning_rate * dm
        b = b - learning_rate * db

    return m, b

# Example Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])  # Perfect fit: y = 2x + 3

final_m, final_b = gradient_descent(X, y)
print(f"Optimized Slope (m): {final_m:.2f}")
print(f"Optimized Intercept (b): {final_b:.2f}")
```
