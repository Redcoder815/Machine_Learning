```python
import numpy as np

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Cost function
def cost_function(y, y_pred):
    m = len(y)
    cost = -(1/m) * np.sum(y * np.log(y_pred) + (1-y) * np.log(1-y_pred))
    return cost

# Example dataset
X = np.array([0.2, 0.4, 0.6])
y = np.array([0, 1, 1])

# Random weights
theta = 0.5 #ùúÉ controls how strongly the input ùë• should influence the prediction. Without multiplying by ùúÉ
, the model cannot learn anything.
z = X * theta
y_pred = sigmoid(z)

print("Predicted Probabilities:", y_pred)
print("Cost Function Value:", cost_function(y, y_pred))
```
